# Project Management in Python

## PIP, Conda, and Virtual Environment

### Python Project includes

- Python version
- Libraries (packages)

### Dependency Conflict

- different project require different versions
- update one package breaks another project

### Pip

- By default installs globally

- Example : pip install pygame,
            pip install numpy

### Why global Installation is a bad idea??

##### Global python is like your sharing a wardrobe with everyone in your house

- Problems:
    - version conflict
    - permission issues
    - hard to reproduce environments
    - difficult to uninstall safely

### Virtual Environments

##### Private python world for one project (your own wardrobe)

- Virtual Environment include:
    - its own python interpreter
    - its own pip
    - its own installed packages

- Each project get:
    - isolation
    - stability
    - reproducibility
