## Tuple packing and unpacking

```python
person = ("Alice", 30, "New York")
name, age, city = person
print(name, age, city)
```

```python
numbers = (1, 2, 3, 4, 5)
first, *middle, last = numbers
print(first)
print(middle)
print(last)
```

```python
data = [4, 1, 9, 2, 7, 5]

maximum = data[0]
minimum = data[0]

for number in data:
    if number > maximum:
        maximum = number
    if number < minimum:
        minimum = number

answer = (maximum, minimum)

print(tuple(answer))
```
