### Parameters

variable listed inside the function definition (placeholders)

- Positional parameters
- Keyword parameters
- Default parameters
- Variable positional parameters (*args)
- Variable keyword parameters (**kwargs)
- Keyword-only parameters

```python
def add(a, b):
    print(a + b)

add(5, 7)
```

```python
def introduce(name, age):
    print(f"My name is {name} and I am {age} years old.")

introduce(name="Bob", age=25)
```

```python
def introduce(name="Arm"):
    print(f"My name is {name}.")

introduce()
introduce("Charlie")
```

```python
def total(*numbers):
    print(numbers)
    print(sum(numbers)) # Variable positional arguments packed into a tuple

total(1, 2, 3, 4, 5, 6)
```

```python
def display_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}") # Keyword arguments packed into a dictionary

display_info(name="Alice", age=30, city="New York")
```

### Arguments
actual values you pass into the function

```python
def greet(name):
    print("Hello", name)

greet("Alice")
```

```python
def introduce(name, age, country="Unknown"):
    print(f"My name is {name} and I am {age} years old. I am from {country}.")

introduce("David", 28)
introduce("Eva", 22, country="Canada")
```

## Return value

```python
def add(a, b):
    return a + b

result = add(3, 5)
print("The sum is:", result)
```

```python
def get_stats(numbers):
    return max(numbers), min(numbers), sum(numbers) // len(numbers)

high, low, average = get_stats([10, 20, 30, 40, 50])
print(high, low, average)
```
