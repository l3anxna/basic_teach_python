# Error Handling

## Syntax error

- your code break the rule of python
- detected before running

```python
print("Hello")
```

## Runtime Error (Exceptions)

- happen while program is running

### Types of exception

- ZeroDivisionError
- ValueError
- TypeError
- IndexError
- KeyError
- FileNotFoundError
- PermissionError
- ImportError

```python
try:
    x = int(input("Enter a number: "))
except:
    print("Invalid input")
```

```python
try:
    x = int(input("Enter a number: "))
    x = 10 / x
except (ValueError, ZeroDivisionError):
    print("Invalid input: Please enter a valid integer.")
```

```python
try:
    x = int(input("Enter a number: "))
except ValueError:
    print("Invalid input: Please enter a valid integer.")
else:
    print(f"You entered: {x}")
```

```python
try:
    f = open("non_existent_file.txt")
    print(f.read())
except FileNotFoundError:
    print("File not found. Please check the file name and try again.")
finally:
    print("Execution completed.")
```

```python
try:
    f = open("non_existent_file.txt")
except FileNotFoundError:
    print("File not found. Please check the file name and try again.")
else:
    print(f.read())
    f.close()
finally:
    print("Execution completed.")
```

## Raising Exceptions

- intentionally trigger error

```python
raise ZeroDivisionError("This is a custom error message.")
```

```python
def set_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative.")
    print(f"Age set to: {age}")

set_age(-5)
```

```python
try:
    x = int("abc")
except ValueError as e:
    print(f"Caught an exception: {e}")
    raise
```

## Creating Custom Exceptions

```python
class NegativeAgeError(Exception):
    pass

def set_age(age):
    if age < 0:
        raise NegativeAgeError("Age cannot be negative.")
    print(f"Age set to: {age}")
```
