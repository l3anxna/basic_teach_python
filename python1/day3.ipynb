```python
import turtle
t = turtle.Turtle()
```

```python
t.forward(100)
t.penup()
t.right(90)
t.forward(100)
t.pendown()
t.forward(100)
```

```python
t.color("yellow")
t.forward(100)
```

```python
t.color("blue")
t.begin_fill()
t.circle(50)
t.end_fill()
```

```python
colors = ["red", "purple", "blue", "green", "orange", "pink"]
t.speed(0)
for i in range(50):
    t.pencolor(colors[i % 6])
    t.forward(i * 5)
    t.right(144)
    
```

```python
t.speed(1)
t.shape("")
t.goto(100, 100)
for _ in range(4):
    t.forward(100)
    t.right(90)
```

```python
import turtle

screen = turtle.Screen()
screen.bgcolor("skyblue")
t = turtle.Turtle()
t.speed(0)

t.penup()
t.goto(200, 150)
t.pendown()
t.color("yellow")
t.begin_fill()
t.circle(50)
t.end_fill()

t.penup()
t.goto(-150, 180)
t.pendown()
t.color("white")
t.begin_fill()
for _ in range(6):
    t.circle(20)
    t.penup()
    t.forward(30)
    t.pendown()
t.end_fill()

t.penup()
t.goto(-50, 220)
t.pendown()
t.begin_fill()
for _ in range(6):
    t.circle(20)
    t.penup()
    t.forward(30)
    t.pendown()
t.end_fill()

t.penup()
t.goto(100, 200)
t.pendown()
t.begin_fill()
for _ in range(6):
    t.circle(20)
    t.penup()
    t.forward(30)
    t.pendown()
t.end_fill()

t.penup()
t.goto(-100, -100)
t.pendown()
t.color("black", "orange")
t.begin_fill()
for _ in range(4):
    t.forward(200)
    t.left(90)
t.end_fill()

t.penup()
t.goto(-100, 100)
t.pendown()
t.color("black", "brown")
t.begin_fill()
t.forward(200)
t.left(120)
t.forward(200)
t.left(120)
t.forward(200)
t.end_fill()

t.penup()
t.goto(-20, -100)
t.pendown()
t.color("black", "blue")
t.begin_fill()
for _ in range(2):
    t.forward(40)
    t.left(90)
    t.forward(80)
    t.left(90)
t.end_fill()

t.penup()
t.goto(-80, 0)
t.pendown()
t.color("black", "lightyellow")
t.begin_fill()
for _ in range(4):
    t.forward(40)
    t.left(90)
t.end_fill()

t.penup()
t.goto(40, 0)
t.pendown()
t.begin_fill()
for _ in range(4):
    t.forward(40)
    t.left(90)
t.end_fill()

t.hideturtle()
```
