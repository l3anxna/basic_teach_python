## String Operation

```python
s1 = "Hello"
s2 = "World"
print(s1 + " " + s2) #concatenation
print(s1 * 4) #repetition
print(len(s1)) #length
print("e" in s2) #membership
print("apples" == "banana") #lexicographical comparison
```

## String Methods

### Changing case

```python
name = "Arm" 
print(name.lower())
print(name.upper())
print(name.capitalize())
```

### Striping whitespace

```python
text = "   Hello World   "
print(text.strip())
```

### Searching and couting

```python
text = "banana"
print(text.find("na"))
print(text.count("ana"))
```

### spliting and joining

```python
sentence = "I love programming"
words = sentence.split()
print(words)

print(" ".join(words))
```

```python
password = input("Enter a password: ")

if len(password) < 8:
    print("Password must be at least 8 characters long.")
elif password.islower():
    print("Password must contain at least one uppercase letter.")
elif password.isupper():
    print("Password must contain at least one lowercase letter.")
elif password.isalpha():
    print("Password must contain at least one digit.")
else:
    print("Password is valid.")
```

```python
password = input("Enter a password: ")

has_upper = False
has_lower = False
has_digit = False

for i in password:
    if i.isupper():
        has_upper = True
    elif i.islower():
        has_lower = True
    elif i.isdigit():
        has_digit = True

if len(password) < 8:
    print("Password must be at least 8 characters long.")
```
