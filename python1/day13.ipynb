# Encapsulation

```python
class Student:
    def __init__ (self, name : str, score : int) -> None:
        self.name = name # public attribute
        self.__score = score # private attribute

    def get_score(self) -> int:
        return self.__score

    def set_score(self, score : int) -> None:
        if 0 <= score <= 100:
            self.__score = score
        else:
            print("Invalid score")
```

```python
s1 = Student("Alice", 85)
print(s1.get_score())
```

# Inheritance

## Base Class / Parent Class

```python
class Person:
    def __init__(self, name : str) -> None:
        self.name = name

    def greet(self) -> str:
        print(f"Hello, my name is {self.name}.")
```

## Child Class

```python
class Student(Person):
    def __init__(self, name : str, score : int) -> None:
        super().__init__(name)
        self.score = score
```

```python
s2 = Student("Bob", 90)
s2.greet()
```

### Only use inheritance for "is-a" relationship

# Polymorphism

```python
class Animal:
    def __init__(self, name : str) -> None:
        pass

    def speak(self) -> str:
        pass
```

```python
class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)
        
    def speak(self) -> str:
        return "Woof!"
```

```python
class Cat(Animal):
    def __init__(self, name):
        super().__init__(name)
        
    def speak(self) -> str:
        return "Meow!"
```

```python
animals = [Dog("Buddy"), Cat("Persia")]

for animal in animals:
    print(animal.speak())
```

Create base class Shape
    -method area
    -method perimeter

Create children class Rectangle(width, height) and Circle(radius)

Each implements its own area and perimeter

```python
class Shape:
    def area(self):
        pass
    def perimeter(self):
        pass
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(self):
        return (self.width + self.height) * 2
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return self.radius**2 * 3.14
    def perimeter(self):
        return (self.radius * 2) * 3.14
shapes = [Rectangle(2, 4), Circle(44)]

for shape in shapes:
    print(shape.area())
    print(shape.perimeter())
```
