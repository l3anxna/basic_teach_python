## Important
### Core Idea
### OOP = Organize code around real-world objects

```python
score = 80
```

- Real-World                Programming
- Blueprint                 Class
- Person                    Object
- Actions                   Methods
- Properties                Attributes

```python
class Student: # PascalCase
    def __init__(self, name : str, score : int) -> None:
        self.name = name  # attribute
        self.score = score  # attribute
    
    def is_passed(self) -> bool: # method
        return self.score >= 75
    
```

```python
s1 = Student("Alice", 85)
s2 = Student("Bob", 70)
s3 = Student("Charlie", 90)

print(s2.is_passed())
print(s3.is_passed())
```

class Cat

- attribute : name, color, energy = 50
- method : meow, eat

```python
class Cat:
    def __init__(self, name = str, color = str, meow = str, eat = str):
        self.name = name # attribute
        self.color = color # attribute
    def meow(self): # method
        print("meow")
    def eat(self): # method
        print("eat")

cat1 = Cat("red", "white")
cat2 = Cat("blue", "black")
cat3 = Cat("green", "white")

cat3.meow()
cat2.eat()
```
