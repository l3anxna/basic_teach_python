# Conditional & Loops

```python
if condition:
    do something
```

```python
age = 20
if age >= 21:
    print('adult')

```

```python
number = 8
if number % 2 == 0:
    print('even')
else:
    print('odd')
```

```python
score = 70
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'F'

print(grade)
```

```python
age = 20
id = True
if age >= 18:
    if id:
        print('admit')
    else:
        print('no admit')
else:
    print('too young')
```

## Loops

## while loop

```python
count = 0
while count <= 5:
    print(count)
```

## for loop

```python
for i in range(1, 10):
    if i == 3:
        continue
    elif i == 8:
        break
    print(i)
```

```python
for i in range(5):
    for j in range(3):
        print(i, j)
```
