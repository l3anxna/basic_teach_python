## Variable scope and lifetime

```python
x = 10 # Global variable

def show():
    x = 5 # Local variable
    print("Inside", x)

show()
print("Outside", x)
```

```python
count = 0

def increment():
    global count
    count += 1

increment()
print("Count after increment:", count)
```

## Lambda function

syntax

lambda arguments: expression

```python
add = lambda x, y: x + y
```

```python
def add(x, y):
    return x + y
```

```python
students = [("John", 88), ("Emma", 92), ("Sam", 85)]
students.sort(key=lambda student: student[1], reverse=True)
print(students)
```

## Recursion

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))
```
