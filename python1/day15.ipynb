```python
import pygame
pygame.init()

screen = pygame.display.set_mode((800, 600))
player = pygame.Rect(400, 250, 50, 50)

running = True

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    keys = pygame.key.get_pressed()
    if keys[pygame.K_a]:
        player.x -= 10
    if keys[pygame.K_d]:
        player.x += 10
    if keys[pygame.K_w]:
        player.y -= 10
    if keys[pygame.K_s]:
        player.y += 10

    if player.x < 0:
        player.x = 0
    if player.x > 750:
        player.x = 750
    if player.y < 0:
        player.y = 0
    if player.y > 550:
        player.y = 550

    screen.fill((255, 255, 255))
    pygame.draw.rect(screen, (255, 0, 0), player)
    pygame.display.update()

pygame.quit()
```
