# File Handling and Context Manager

```python
open("filename", "mode")
```

File Modes
- "r" for read only
- "w" for write (overwrite)
- "a" for append
- "r+" for read and write

## Reading file

```python
file = open("example.txt", "r+")
file.write("Good morning")
file.seek(0)
print(file.read())
file.close()
```

```python
file = open("example.txt", "r")
content = file.read()
print(content)
file.close()
```

```python
file = open("example.txt", "r")

for line in file:
    print(line)

file.close()
```

```python
file = open("example.txt", "r")
file.seek(3)
line = file.readline()
print(line)
file.close()
```

## Writing file

```python
file = open("example.txt", "w")
file.write("New content")
file.close()
```

```python
file = open("example.txt", "a")
file.write("Hello World!\n")
file.close()
```

## Context Manager

```python
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
with open("example.txt", "r+") as file:
    file.write("Trying to write in read mode")
```
